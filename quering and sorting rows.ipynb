```python
import pandas as pd

covid_df = pd.read_csv('italy-covid-daywise.csv')
print(covid_df)
```

```python
# Querying and sorting rows

#Let's say we want to look at the days which had more than 1000 reported cases. 
#We can use a boolean expression to check which rows satisfy this criterion.
more1000Cases = covid_df.new_cases>1000
# print(more1000Cases) # series of boolean
# print(covid_df[more1000Cases])

# in one line
# print(covid_df[covid_df.new_cases > 1000])
# by default jupyter print first and last five rows
# to print all rows
print(covid_df[covid_df.new_cases > 1000].to_string()) 
# second way by importing display module

# from IPython.display import display

# with pd.option_context('display.max_rows',100):
#     print(display(covid_df[covid_df.new_cases>1000]))

```

```python
# adding new column


covid_df['positiveRate'] = covid_df.new_cases / covid_df.new_tests
print(covid_df['positiveRate'].to_string())
```

```python
# print(covid_df)
# How to delete a colmn
covid_df.drop(columns=['positiveRate'], inplace=True)
print(covid_df)
```
